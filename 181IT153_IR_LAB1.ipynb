```python
import pandas as pd
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
from collections import defaultdict
```

```python
# nltk.download('wordnet')
```

```python
corpus=pd.read_csv("historic_articles.csv",encoding='latin-1', usecols=["content"], nrows=100)
corpus=corpus["content"].to_list()
corpus
```

```python
# Step 1: Tokenization-ANALYZING INITIAL CORPUS
tokenized_corpus=[]
initial_vocabulary_count=0
for document in corpus:
    tokenized_corpus.append(nltk.word_tokenize(document))
    initial_vocabulary_count+=len(tokenized_corpus[len(tokenized_corpus)-1])
print("Initial vocabulary size:-",initial_vocabulary_count)
```

```python
# Step 2: Normalization 
punctuations= '''!()[]{}.-';:"\, <>/?@#$%^&*_~'''

i=0
for document in corpus:
    for word in document:
        if word in punctuations:  
            document = document.replace(word, " ")            
    document=' '.join(document.split())
    document=document.lower()
    corpus[i]=document
    i+=1


```

```python
# LEMMATIZATION  
lmtzr = WordNetLemmatizer()
i=0
for document in corpus:
    current_doc=""
    for word in document.split():
        if(word in stopwords.words()):
            current_doc+=""
        else:
            current_doc+=lmtzr.lemmatize(word)+" "
    corpus[i]=current_doc
    i+=1
```

```python
vocabulary_size_after_preprocessing=0
for document in corpus:
    vocabulary_size_after_preprocessing+=len(document.split())
print("Vocabulary Size post the Preprocessing is",vocabulary_size_after_preprocessing)
```

```python
# Generating term postings
term_postings=[]
doc_id=1
for document in corpus:
    for word in document.split():
        term_postings.append([word,doc_id])
    doc_id+=1
term_postings.sort()
term_postings
```

```python
posting_list=defaultdict(lambda:0)
for terms in term_postings:
    posting_list[terms[0]]+=1
posting_list
```

```python
inverted_index=dict()
term_postings=list(set(map(tuple,term_postings)))
term_postings.sort()
term_postings
for key,value in posting_list.items():
    temp=[]
    for i in term_postings:
        if(i[0]>str(key)):
            break
        elif(i[0]==str(key)):
            temp.append(i[1])
    inverted_index[key]={value:temp}
inverted_index
```

```python
term1="parliament"
term2="delhi"
term3="india"

if term1 in inverted_index.keys():
    set1=set(list(inverted_index[term1].values())[0])

if term2 in inverted_index.keys():
    set2=set(list(inverted_index[term2].values())[0])

if term3 in inverted_index.keys():
    set3=set(list(inverted_index[term3].values())[0])
print("Query 1 constst of:",term1,"and",term2,"and",term3)
print("Document ID's which are relevant",set1&set2&set3)
```

```python
print("Query 2 :",term1,"or",term2,"but not",term3)
print("Relevant Document ID's are",(set1|set2) - set3)
```
